```python
#unit step signal

import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-10,11,1)
unit = np.where(n>=0,1,0)

plt.scatter(n,unit)
plt.stem(n,unit,label = "Unit Step ")

plt.title("Unit Step Signal")
plt.xlabel("n")
plt.ylabel("Unit")
plt.legend()

plt.grid()

plt.show()
```

```python
#Ramp signal

import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-10,10,1)
ramp = np.where(n>=0,n,0)
plt.stem(n,ramp,label="u(n)")

plt.title("Ramp signal")
plt.xlabel("n")
plt.ylabel("ramp")
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(0, 20)
a = 0.9
x = a ** n

plt.stem(n, x)
plt.title('Discrete-Time Exponential Signal')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid(True)
plt.tight_layout()
plt.show()
```
