```python
import numpy as np
import matplotlib.pyplot as plt

# Define the analog signal
def analog_signal(t):
    return 3 * np.cos(2000 * np.pi * t) + 5 * np.sin(6000 * np.pi * t) + 10 * np.cos(12000 * np.pi * t)

# Time duration
t_max = 0.002  # 2 ms, enough to show several cycles of the highest frequency (6000 Hz)

# Define sampling frequencies
fs1 = 15000  # No aliasing
fs2 = 8000   # Aliasing occurs (less than 2 * 6000 Hz)

# Sampled time vectors
t1 = np.arange(0, t_max, 1/fs1)
t2 = np.arange(0, t_max, 1/fs2)

# Sampled signals
x1 = analog_signal(t1)
x2 = analog_signal(t2)

# Plotting
plt.figure(figsize=(12, 6))

# Plot for fs1
plt.subplot(2, 1, 1)
plt.stem(t1, x1)
plt.title("Sampled Signal at 15000 Hz (No Aliasing)")
plt.ylabel("x[n]")
plt.grid(True)

# Plot for fs2
plt.subplot(2, 1, 2)
plt.stem(t2, x2)
plt.title("Sampled Signal at 8000 Hz (Aliasing Occurs)")
plt.xlabel("Time (s)")
plt.ylabel("x[n]")
plt.grid(True)

plt.tight_layout()
plt.show()
```
